```python
import yfinance as yf
import plotly.graph_objects as go

ticker = yf.Ticker('XOM').info
market_price = ticker['regularMarketOpen']
previous_close_price = ticker['regularMarketPreviousClose']
print('Ticker: XOM')
print('Market Price:', market_price)
print('Previous Close Price:', previous_close_price)
```

# Candlestick and moving average below

```python
tickerlist = 'QQQ'

data = yf.download(tickers=tickerlist, period='120mo', interval='1wk')

fig = go.Figure(data=[go.Candlestick(x=data.index,
                                     open=data['Open'],
                                     high=data['High'],
                                     low=data['Low'],
                                     close=data['Close'])])

fig.show()

data['SMA30'] = data['Close'].rolling(50).mean()
data['Open'].plot(label='Open', color='orange')
data['SMA30'].plot(label='SMA')

#print(ticker)
```
